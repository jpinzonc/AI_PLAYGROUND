```python
from langchain_community.chat_models import ChatOllama
# This it the language model we'll use. We'll talk about what we're doing below in the next section
chat = ChatOllama(model="llama3")
```

### CHATS 

```python
from langchain.schema import HumanMessage, SystemMessage, AIMessage
chat(
    [
        SystemMessage(content="You are a nice AI marine biologist that helps a user figure out ocean facts in one short sentence"),
        HumanMessage(content="Are sharks dangerous?")
    ]
)
```

```python
# compiled trom https://python.langchain.com/v0.2/docs/integrations/chat/ollama/
from langchain.schema import HumanMessage, SystemMessage, AIMessage
messages =     [
        SystemMessage(content="You are a marine biologist that can offer science insights in 500 words or less"),
        HumanMessage(content="Are sharks dangerous?")
    ]

chat_model_response = chat.invoke(messages)
print(chat_model_response)
```

```python
print(chat_model_response.content)
```
